```python
from modeling import SRNsModel
import util

_MODEL_PATH = '/mnt/wmy/liury/srn/logs/040117body_part/checkpoints/epoch_4204_iter_370000.pth'

_NUM_INSTANCES = 101

model = SRNsModel(num_instances=_NUM_INSTANCES,
                  latent_dim=256,
                  tracing_steps=10,
                  freeze_networks=True,
                  out_channels=16,
                  img_sidelength=128,
                  output_sidelength=512)

util.custom_load(model, path=_MODEL_PATH, discriminator=None,
                 overwrite_embeddings=False)

model.eval()
model.cuda()
```

```python
import os
from glob import glob
from torch.utils.data import DataLoader
from dataset.body_dataset import BodyPartDataset

_DATA_ROOT = '/data/anpei/facial-data/seg_body'

_OUTPUT_DIR = '/mnt/wmy/liury/srn/logs/040117body_part/output'

_NUM_INTERP = 8

os.makedirs(_OUTPUT_DIR, exist_ok=True)

dataset = BodyPartDataset(
            root_dir=_DATA_ROOT, 
            data_type='seg',
            img_sidelength=128,
            sample_observations=list(range(8)))

print('[DONE] load val dataset.', len(dataset))

cam_poses = []

for sample in dataset:
    print(sample[0][0].keys())

dataloader = DataLoader( dataset,
                         collate_fn=dataset.collate_fn,
                         batch_size=_NUM_INTERP,
                         shuffle=False,
                         drop_last=False)
```

```python
import torch
import numpy as np

src_idx = torch.randint(0, _NUM_INSTANCES, (1,)).squeeze().cuda()
trgt_idx = src_idx + 5

print('sampled frames = ', src_idx, trgt_idx)

z_src = model.get_embedding({'instance_idx': src_idx}).unsqueeze(0).repeat(_NUM_INTERP, 1)
z_trgt = model.get_embedding({'instance_idx': trgt_idx}).unsqueeze(0).repeat(_NUM_INTERP, 1)

interp = torch.Tensor(np.linspace(0.0, 1.0, _NUM_INTERP)).squeeze().cuda().unsqueeze(1)
z_interp = z_src * (1.0 - interp) + z_trgt * interp

print('z_interp = ', z_interp.shape)
```

```python
import torch
from torchvision.utils import make_grid
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import numpy as np

matplotlib.rcParams['figure.figsize'] = [30, 5]

with torch.no_grad():
        
    for idx, model_input in enumerate(dataloader):
        
        if idx > 0:
            break
        
        print('******************************* %03d *******************************'%(idx))
        
        model_input, _ = model_input
        
        pose = model_input['pose']
        intrinsics = model_input['intrinsics']
        
        uv = model_input['uv']
#         if 'params' in model_input.keys() and model_input['params'] is not None:
#             z = model_input['params']
#         else:
#             z = model.get_embedding(model_input)

#         print(pose.shape, uv.shape)

        for v_idx in range(pose.shape[0]):
            cur_pose = pose[v_idx:v_idx+1, ...].repeat(_NUM_INTERP, 1, 1)
            cur_intr = intrinsics[v_idx:v_idx+1, ...].repeat(_NUM_INTERP, 1, 1)
            cur_uv = uv[v_idx:v_idx+1, ...].repeat(_NUM_INTERP, 1, 1)

            predictions, depth_map = model(cur_pose, z_interp, cur_intr, cur_uv)

            B, _, C = predictions.shape

            pred = torch.argmax(predictions, dim=2, keepdim=True)
            output_img = util.lin2img(pred, color_map=dataset.color_map)
            output_img = make_grid(output_img, nrow=8, padding=1).permute((1, 2, 0)).cpu().numpy()

            print(idx, output_img.shape, type(output_img), np.max(output_img))

            plt.imshow(output_img)
            plt.show()
```
